# Task 1: Conceptual Understanding




1. What is the difference between "Love" and "love" in NLP?

In NLP, "Love" and "love" are treated as different tokens if the text is case-sensitive. This means the model may treat them as two different words, which increases vocabulary size and reduces model efficiency. Converting all text to lowercase ensures that words like "Love", "LOVE", and "love" are treated as the same token, improving consistency and model performance.

2. What happens if stopwords are not removed?

If stopwords are not removed:
The dataset will contain many common words like is, the, am, are which do not carry significant meaning.
This increases processing time and memory usage.
Machine learning models may focus on unimportant words, reducing accuracy.

3. Provide two real-world scenarios where removing stopwords can be harmful.

Removing stopwords is not always good. Sometimes stopwords carry important meaning.

a..Sentiment Analysis
Sentence: "I am not happy"
If we remove stopword "not", sentence becomes "I happy" which changes the meaning completely.

b..Question Answering Systems
Sentence: "What is the time?"
Removing stopwords may remove "what", which is important for understanding the question.

4. Explain the Difference between Stemming and Lemmatization.

Stemming:
Stemming reduces a word to its root form by removing suffixes using simple rules.
It is fast but may produce incorrect or non-meaningful words (e.g., studies → studi).

Lemmatization:
Lemmatization converts a word to its base dictionary form using grammar and context.
It is more accurate and always produces meaningful words (e.g., studies → study).


# Task 2:Build Advanced Preprocessing Function

```python
import re

def preprocess_text(text):
    # 1. Convert to lowercase (Solves the "Love" vs "love" issue)
    text = text.lower()

    # 2. Remove URLs and Email-like patterns
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'\S+@\S+\.\S+', '', text)

    # 3. Handle repeated characters (e.g., "soooo goooood" -> "so good")
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 5. Remove punctuation (to clean up things like "!!!")
    text = re.sub(r'[^\w\s]', '', text)

    # 6. Tokenize and Filter (Short tokens & Stopwords)
    exceptions = {'no', 'not'}
    tokens = text.split()

    clean_tokens = [
        word for word in tokens
        if len(word) > 2 or word in exceptions
    ]

    # 7. Remove extra spaces and rejoin
    return " ".join(clean_tokens).strip()


test_cases = [
    "I have 2 dogs",
    "This   is      good",
    "soooo goooood!!!",
    "WOW!!! This is GREAT!!!",
    "Visit http://nandishgs0009@gmail.com",
    "mail me@test.com",
    "I said no to the cat"
]

print(f"{'INPUT':<50} | {'OUTPUT'}")
print("-" * 80)
for tc in test_cases:
    print(f"{tc:<50} | {preprocess_text(tc)}")
```

```python
import re

def preprocess_text(text):
    # 1. Lowercase
    text = text.lower()

    # 2. URLs & Email Patterns
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'\S+@\S+\.\S+', '', text)

    # 3. Repeated Characters (3 or more become 1)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 4. Numbers
    text = re.sub(r'\d+', '', text)

    # 5. Special characters/Emojis/Punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # 6. Tokenize & Short Token Filter (with "no/not" exception)
    exceptions = {'no', 'not'}
    tokens = text.split()

    clean_tokens = [word for word in tokens if len(word) > 2 or word in exceptions]

    # 7. Reconstruct Cleaned Sentence
    cleaned_sentence = " ".join(clean_tokens).strip()

    return clean_tokens, cleaned_sentence

sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

print("="*110)
print(f"{'No':<5}{'Original Text':<40}{'Cleaned Tokens':<35}{'Cleaned Sentence'}")
print("="*110)

for i, sentence in enumerate(sample_inputs, 1):
    tokens, cleaned = preprocess_text(sentence)
    print(f"{i:<5}{sentence:<40}{str(tokens):<35}{cleaned}")

print("="*110)
```

# Task 4: Token Analytics

```python
import re

def get_analytics(tokens):
    if not tokens:
        return 0, 0, 0.0

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(t) for t in tokens) / total_tokens

    return total_tokens, unique_tokens, round(avg_length, 2)

def preprocess_and_analyze(text):

    text = text.lower()
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)

    exceptions = {'no', 'not'}
    tokens = [w for w in text.split() if len(w) > 2 or w in exceptions]

    total, unique, avg_len = get_analytics(tokens)

    return {
        "tokens": tokens,
        "total": total,
        "unique": unique,
        "avg_len": avg_len
    }

sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Output Table
print(f"{'Original Sentence':<35} | {'Tokens':<5} | {'Unique':<6} | {'Avg Len'}")
print("-" * 70)

for sentence in sample_inputs:
    data = preprocess_and_analyze(sentence)
    print(f"{sentence[:34]:<35} | {data['total']:<5} | {data['unique']:<6} | {data['avg_len']}")
```

# Task 5: Frequency Analysis

```python
from collections import Counter
import re

def clean_and_tokenize(text):
    text = text.lower()
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)

    exceptions = {'no', 'not'}
    tokens = [w for w in text.split() if len(w) > 2 or w in exceptions]
    return tokens

sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]


all_cleaned_tokens = []
for sentence in sample_inputs:
    tokens = clean_and_tokenize(sentence)
    all_cleaned_tokens.extend(tokens)


word_counts = Counter(all_cleaned_tokens)


top_10 = word_counts.most_common(10)
least_5 = sorted(word_counts.items(), key=lambda x: x[1])[:5]

print("TOP 10 MOST FREQUENT WORDS")
for word, count in top_10:
    print(f"{word:<12}    {count} times")

print("\n TOP 5 LEAST FREQUENT WORDS")
for word, count in least_5:
    print(f"{word:<12}    {count} time")
```

# Task 6: Build Full Pipeline

```python
import re

def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    exceptions = {'no', 'not'}

    for text in text_list:

        text = text.lower()


        text = re.sub(r'https?://\S+|www\.\S+', '', text)
        text = re.sub(r'\S+@\S+\.\S+', '', text)


        text = re.sub(r'(.)\1{2,}', r'\1', text)

        text = re.sub(r'\d+', '', text)
        text = re.sub(r'[^\w\s]', '', text)


        current_tokens = [
            word for word in text.split()
            if len(word) > 2 or word in exceptions
        ]


        clean_sentence = " ".join(current_tokens).strip()


        all_tokens.extend(current_tokens)
        if clean_sentence:
            clean_sentences.append(clean_sentence)


    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


sample_data = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

output = full_pipeline(sample_data)


print("CLEANED SENTENCES")
for i, s in enumerate(output['clean_sentences'], 1):
    print(f"{i}. {s}")

print("\n TOKEN PREVIEW (First 15)")
print(output['tokens'][:15])
```

# Task 7: Error Handling

```python
import re

def full_pipeline_with_error_handling(text_list):

    if not isinstance(text_list, list):
        return {"error": "Input must be a list of strings."}

    all_tokens = []
    clean_sentences = []
    exceptions = {'no', 'not'}

    for text in text_list:

        if not text or not str(text).strip():
            continue  # Skip this empty entry


        text = str(text).lower()


        text = re.sub(r'https?://\S+|www\.\S+', '', text)
        text = re.sub(r'\S+@\S+\.\S+', '', text)

        text = re.sub(r'(.)\1{2,}', r'\1', text)

        text = re.sub(r'\d+', '', text)
        text = re.sub(r'[^\w\s]', '', text)

        current_tokens = [
            word for word in text.split()
            if len(word) > 2 or word in exceptions
        ]

        if not current_tokens:
            continue

        clean_sentence = " ".join(current_tokens).strip()

        all_tokens.extend(current_tokens)
        clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences,
        "processed_count": len(clean_sentences)
    }

edge_cases = [
    "",
    "   ",
    "🚀🚀🚀😍😍",
    "1234567890",
    "!!! ??? @#$",
    "I am not happy"
]

result = full_pipeline_with_error_handling(edge_cases)

print(f"Total Sentences Input: {len(edge_cases)}")
print(f"Total Sentences Successfully Processed: {result['processed_count']}")
print(f"Final Clean Sentences: {result['clean_sentences']}")
```
